```python
import pandas as pd

llama3_1 = pd.read_csv('../../gym/results/llama3_1/llm_response.csv')
gemini1_5 = pd.read_csv('../../gym/results/gemini1_5/llm_response.csv')
gemma2 = pd.read_csv('../../gym/results/gemma2/llm_response.csv')
gpt4o = pd.read_csv('../../gym/results/gpt4o/llm_response.csv')

gemini1_5.drop(columns=['table_body', 'caption', 'references', 'explanation', 'keywords'], inplace=True)
llama3_1.drop(columns=['table_body', 'caption', 'references', 'explanation', 'keywords'], inplace=True)
gpt4o.drop(columns=['table_body', 'caption', 'references', 'explanation', 'keywords'], inplace=True)

llama3_1['model'] = 'llama3_1'
gemini1_5['model'] = 'gemini1_5'
gemma2['model'] = 'gemma2'
gpt4o['model'] = 'gpt4o'

```

```python
#put all the data together and add a column to identify the model
all_llm = pd.concat([llama3_1, gemini1_5, gemma2, gpt4o])
all_llm
```

```python

all_llm.to_csv('../../gym/results/llm_response.csv', index=False)
```
